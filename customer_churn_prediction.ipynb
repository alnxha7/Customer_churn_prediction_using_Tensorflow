```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
df.drop('customerID', axis= 1, inplace= True)
df.head()
```

```python
df.dtypes
```

```python
pd.to_numeric(df.TotalCharges, errors= 'coerce').isnull().sum()
```

```python
df[pd.to_numeric(df.TotalCharges, errors= 'coerce').isnull()]
```

```python
df1 = df[df.TotalCharges != ' ']
df1.shape
```

```python
df1[pd.to_numeric(df1.TotalCharges, errors= 'coerce').isnull()]
```

```python
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)
```

```python
df1.TotalCharges.dtypes
```

```python
tenure_churn_no = df1[df1.Churn == 'No'].tenure
tenure_churn_yes = df1[df1.Churn == 'Yes'].tenure
```

```python
plt.xlabel('tenure')
plt.ylabel('no of customers')
plt.title('customers churn pediction visualization')

plt.hist([tenure_churn_no, tenure_churn_yes], color= ['black', 'green'], label= ['Churn=No', 'Churn=Yes'])
plt.legend()
```

```python
df1.replace('No internet service', 'No', inplace= True)
df1.replace('No phone service', 'No', inplace= True)
df1.head()
```

```python
def print_unique_col_values(df):
    for column in df:
        if df[column].dtypes == 'object':
            print(f'{column}: {df[column].unique()}')
```

```python
print_unique_col_values(df1)
```

```python
df1.columns
```

```python
yes_no_cols = ['Partner', 'Dependents',
       'PhoneService', 'MultipleLines', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'PaperlessBilling', 'Churn']

for col in yes_no_cols:
    df1[col].replace({'Yes': 1, 'No': 0}, inplace= True)
```

```python
print_unique_col_values(df1)
```

```python
for i in df1:
    print(f'{i}: {df1[i].unique()}')
```

```python
df1['gender'].replace({'Female': 1, 'Male': 0}, inplace= True)
df1['gender'].unique()
```

```python
df2 = pd.get_dummies(data= df1, columns= ['InternetService', 'Contract', 'PaymentMethod'], dtype= 'uint8')
print(df2.columns)
```

```python
df2.sample(7)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
cols_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.sample(3)
```

```python
df2.columns
```

```python
X = df2.drop('Churn', axis= 1)
y = df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 5)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20, input_shape= (26,), activation= 'relu'),
    keras.layers.Dense(15, activation= 'relu'),
    keras.layers.Dense(1, activation= 'sigmoid')
])

model.compile(optimizer= 'adam',
             loss= 'binary_crossentropy',
             metrics= ['accuracy'])


model.fit(X_train, y_train, epochs= 50)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_test[:5]
```

```python
y_pred = []

for i in yp:
    if i >= 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, y_pred))
```

```python
import seaborn as sns

cm = tf.math.confusion_matrix(labels= y_test, predictions= y_pred)
plt.figure(figsize= (10, 7))
sns.heatmap(cm, annot= True, fmt= 'd')
plt.xlabel('predicted')
plt.ylabel('truth')
```
